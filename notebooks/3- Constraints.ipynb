```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_excel('DATA.xlsx')
df.head()
```

```python
df.to_csv('data.csv', index=False)
pd.read_csv('data.csv').dtypes
```

```python
agg_df = df.groupby(['Carrier', 'Shipping Date']).count()[['Status']]
agg_df
```

```python
ls = []
for i in set(list(agg_df.index.get_level_values(0))):
    print(agg_df.loc[i,:].max().values[0])
    ls.append({'carrier':i, 'num_orders':agg_df.loc[i,:].max().values[0]})
```

```python
demand_capacity = pd.DataFrame(ls).set_index('carrier')
demand_capacity
```

```python
weight_capacity = df[['Carrier','Weight']].groupby('Carrier').max()
weight_capacity
```

```python
demand_capacity.join(weight_capacity)
```

```python
ls = []
for i in list(df.Carrier.value_counts().index):
    print(', '.join(list(df[df['Carrier'] == i]['City'].unique())))
    ls.append({'carrier':i,
               'Riyadh': int('Riyadh' in list(df[df['Carrier'] == i]['City'].unique())),
               'Dammam': int('Dammam' in list(df[df['Carrier'] == i]['City'].unique())),
               'Madinah': int('Madinah' in list(df[df['Carrier'] == i]['City'].unique())),
               'Jeddah': int('Jeddah' in list(df[df['Carrier'] == i]['City'].unique())),
               'Makkah': int('Makkah' in list(df[df['Carrier'] == i]['City'].unique())),
              })
```

```python
availability = pd.DataFrame(ls).set_index('carrier')
availability
```

```python
demand_capacity.join(weight_capacity).join(availability).to_csv('constraints.csv')
```
